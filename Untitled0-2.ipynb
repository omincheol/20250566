```python
import matplotlib.pyplot as plt

x_values = [1, 2, 3, 4, 5, 6]
probabilities = [1/6 for _ in x_values]

plt.figure(figsize=(8, 6))
plt.stem(x_values, probabilities, basefmt=" ")
plt.plot(x_values, probabilities, 'o', color='blue')
plt.ylim(0, 1)
plt.title('Probability Mass Function of a Fair Die')
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.show()
```

```python
# !pip install scipy

import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt

n_values = [10, 20, 40, 100]
p = 0.5

plt.figure(figsize=(14, 10))

for i, n in enumerate(n_values, start=1):
    x = np.arange(0, n+1)
    pmf_values = stats.binom.pmf(x, n, p)
    plt.subplot(2, 2, i)
    plt.stem(x, pmf_values, linefmt='C0-', markerfmt='C0o', basefmt=" ")
    plt.title(f'{n} Coin Flips')
    plt.xlabel('Number of Heads')
    plt.ylabel('Probability')
    plt.ylim(0, max(pmf_values) + 0.05)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import poisson

lambda_val = 12
k_values = np.arange(0, 31)
pmf_values = poisson.pmf(k_values, lambda_val)

plt.figure(figsize=(10, 6))
markerline, stemlines, baseline = plt.stem(k_values, pmf_values, basefmt=" ")
plt.setp(markerline, color='green', marker='o')
plt.setp(stemlines, color='gray')
plt.setp(baseline, visible=False)
plt.title('Poisson PMF (=12) : Daily Library Visitors')
plt.xlabel('Visitors (k)')
plt.ylabel('Probability P(X=k)')
plt.ylim(0, max(pmf_values) + 0.02)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

p = 0.1
k_values = np.arange(1, 101)
pmf_values = (1 - p) ** (k_values - 1) * p

plt.figure(figsize=(12, 8))
markerline, stemlines, baseline = plt.stem(k_values, pmf_values, basefmt=" ")
plt.setp(markerline, color='purple', marker='o')
plt.setp(stemlines, color='gray')
plt.setp(baseline, visible=False)
plt.title('Geometric PMF: Time to First Cyber Attack')
plt.xlabel('Days Until First Successful Attack')
plt.ylabel('Probability')
plt.ylim(0, 1)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
